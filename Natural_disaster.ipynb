```python
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
extreme_climate_change_sample = extreme_climate_change.head(100)
```

```python
Z = linkage(extreme_climate_change_sample, 'average')
plt.figure(figsize=(75, 30))
D = dendrogram(Z=Z, orientation="right", leaf_font_size=30, labels = extreme_climate_change_sample.index)
```

find kMean using 3 cluster

```python
model = KMeans(n_clusters = 3)
model.fit(extreme_climate_change)
```

```python
model.labels_
```

It shows that it cluster in years basis.

```python
fig = plt.figure(figsize=(5, 5))
 
# Create a colormap
colormap = np.array(['yellow', 'red', 'blue'])
 
plt.scatter(extreme_climate_change.Temp, extreme_climate_change.Year, c=colormap[model.labels_], edgecolor = 'black', s = 80)
plt.title('K Mean Classification')
plt.xlabel('Temp')
plt.ylabel('Year')
```
